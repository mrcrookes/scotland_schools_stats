```python
# Import key librarys
import pandas as pd
import numpy as np
import os
import glob
```

```python
# use glob to get all the csv files in the folder
path = './highland_primary/'
csv_files = glob.glob(os.path.join(path, "*.csv"))

# Create list for all dfs
frames = []

for f in csv_files:
    
    wdf = pd.read_csv(f)
    
    frames.append(wdf)
```

```python
wdf.head()
```

```python
## Concat all together long format dfs into final df
ldf = pd.concat(frames)
```

```python
# Rename percentage categories so that they are numbers
ldf['value'] = ldf['value'].replace(
    {'0-10%':'5%',
     '10-20%':'15%',
     '20-30%':'25%',
     '30-40%':'35%',
     '40-50%':'45%',
     '50-60%':'55%',
     '60-70%':'65%',
     '70-80%':'75%',
     '80-90%':'85%',
     '90-100%':'95%',
    })

# Rename school names
ldf['school_name'] = ldf['school_name'].replace(
    {'Bun-Sgoil Ghàidhlig Phort Rìgh':'Bun-sgoil Ghaidhlig Phort Righ',
     'Bun-sgoil Ghaidhlig Inbhir Nis':'Bun-sgoil Ghaidhlig Inbhir Nis',
     'Bun-Sgoil Shleite':'Bun-sgoil Shleite',
     'Bun-sgoil Ghàidhlig Loch Abar':'Bun-sgoil Ghaidhlig Loch Abar',
    })

# Drop duplicates
ldf = ldf.drop_duplicates()
```

```python
ldf.to_csv('highland_primary.csv')
```
